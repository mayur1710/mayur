```python
import pandas as pd
```

```python
#Q2.Find out the names of passengers younger than 35 years.

# read the dataset into a pandas dataframe
train_df = pd.read_csv("train.csv")

young_passengers = train_df.loc[train_df['Age'] < 35, 'Name']

for i, name in enumerate(young_passengers, start=1):
    print(f"{i} {name}")


```

```python
# Q4. Find out the statistics aggregate of Age & Fare using the DataFrame.agg() method

train_df = pd.read_csv("train.csv")

# select the Age and Fare columns
age_fare_df = train_df[['Age', 'Fare']]

agg_stats = age_fare_df.agg(['count', 'mean', 'median', 'min', 'max', 'std'])

# print the statistics aggregate
print(agg_stats)
```

```python
#Q5.Find out the mean ticket fare price for each of the sex and cabin class combinations


train_df = pd.read_csv("train.csv")

fare_by_sex_cabin = train_df.groupby(['Sex', 'Cabin'])['Fare'].mean()

print(fare_by_sex_cabin)
```
